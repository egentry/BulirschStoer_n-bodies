```python
body_1_filename_new = 'body_001.dat'
body_2_filename_new = 'body_002.dat'
body_3_filename_new = 'body_003.dat'

t_1_new, x_1_new, y_1_new, z_1_new = np.loadtxt(body_1_filename_new, unpack=True, skiprows=1)
t_2_new, x_2_new, y_2_new, z_2_new = np.loadtxt(body_2_filename_new, unpack=True, skiprows=1)
t_3_new, x_3_new, y_3_new, z_3_new = np.loadtxt(body_3_filename_new, unpack=True, skiprows=1)

def plot_solution(plot_start=40):
    plot_step_size = 10
    print("plot range: t= ",plot_start,  " - ", plot_start + plot_step_size)

    indices_new = np.where( ((t_1_new > plot_start) & (t_1_new < plot_start + plot_step_size)))
    plt.plot(x_1_new[indices_new], y_1_new[indices_new], linestyle='-')
    plt.plot(x_2_new[indices_new], y_2_new[indices_new], linestyle='dotted')
    plt.plot(x_3_new[indices_new], y_3_new[indices_new], linestyle='--')
    plt.xlabel(r"x")
    plt.ylabel(r"y")
    
    plt.legend((r"Body 1", r"Body 2", r"Body 3"))
    
    return

interact(plot_solution, plot_start=(0,60,10));
```
